```python
# Remove all empty lines in a text-file here

input_file = "correct_text.txt"
with open(input_file, "r", encoding="utf-8") as f:
    lines = [l for l in f if l.strip()]

with open(input_file, "w", encoding="utf-8") as f:
    f.writelines(lines)
```

```python
#This code block is just for testing

from fuzzywuzzy import fuzz

file1 = 'wrong_text.txt'
file2 = 'fixed_text.txt'

with open(file1, "r", encoding="utf-8") as f:
    lines1 = [l for l in f if l.strip()]

with open(file2, "r", encoding="utf-8") as f:
    lines2 = [l for l in f if l.strip()]

for i in range(len(lines1)):
    fuzzy_score = fuzz.ratio(lines1[i], lines2[i])
    # print(fuzzy_score)
    if (fuzzy_score < 50):
        print(i, "---------")
        print(lines1[i])
        print(lines2[i])


```
